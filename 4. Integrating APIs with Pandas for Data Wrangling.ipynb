<h2> Exploring HTML Pages

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')


# Extracting paragraphs and creating a DataFrame
wikipedia_paragraphs = []
for p in soup.find_all('p'):
    text = p.get_text().strip()  # strip() removes leading and trailing whitespace
    if text:  # Only append non-empty strings
        wikipedia_paragraphs.append(text)
    df_paragraphs = pd.DataFrame(wikipedia_paragraphs)
print(df_paragraphs)
```

```python
df_paragraphs.shape[0]
```

```python
link_texts = []
link_hrefs = []
for link in soup.find_all('a'):
    link_texts.append(link.get_text().strip())
    link_hrefs.append(link.get('href'))

links_dict = {'Link Text': link_texts, 'URL': link_hrefs}
df_links = pd.DataFrame(links_dict)
df_links
```

```python
import pandas as pd
import pycountry
import pycountry_convert as pc

# Read file csv
file_path = 'df_wikipedia.csv'  
df_wikipedia = pd.read_csv(file_path)

def get_continent(country_name):
    try:
        country = pycountry.countries.get(name=country_name)
        if not country:
            country = pycountry.countries.get(alpha_3=country_name)
        if not country:
            return None
        country_alpha2 = country.alpha_2
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except Exception as e:
        return None

# Add column "Continent"
df_wikipedia['Continent'] = df_wikipedia['Location'].apply(get_continent)
df_wikipedia
```

```python
countries_pop_over_10000 = df_wikipedia[df_wikipedia["Population"] > 10000]
countries_pop_over_10000 = countries_pop_over_10000[countries_pop_over_10000['Location'] != 'World']
continent_population = countries_pop_over_10000.groupby('Continent')['Population'].mean()
continent_population.plot(kind='bar')
plt.ylabel('Mean population')
plt.title("Mean population by Continent")
plt.show()
```
